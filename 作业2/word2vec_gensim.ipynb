```python
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec
```

```python
# 读入训练集文件
data = pd.read_csv('train.csv')
```

```python
# 转字符串数组
corpus = data['comment'].values.astype(str)
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                          .replace("，", "")
                          .replace("!", "")
                          .replace("！", "")
                          .replace("。", "")
                          .replace("~", "")
                          .replace("；", "")
                          .replace("？", "")
                          .replace("?", "")
                          .replace("【", "")
                          .replace("】", "")
                          .replace("#", "")
                        ) for index in range(len(corpus))]
```

```python
# 词向量模型训练
model = Word2Vec(corpus, sg=0, vector_size=300, window=5, min_count=3, workers=4)
#模型显示
print('模型参数：',model,'\n')
```

```python
#最匹配
print('最匹配的词是：',model.wv.most_similar(positive=['点赞', '不错'], negative=['难吃']),'\n')
```

```python
#最不匹配
print('最不匹配的词是：',model.wv.doesnt_match("点赞 好吃 支持 难吃".split()),'\n')
```

```python
#语义相似度
print('相似度为=',model.wv.similarity('推荐','好吃'),'\n')
```

```python
#坐标返回
print(model.wv.__getitem__('地道'))
```

```python
# 任务1：使用Skip-gram模型训练
model2 = Word2Vec(corpus, sg=1, vector_size=300, window=5, min_count=3, workers=4)
```

```python
# 任务2：“环境”的词向量和形状
print("环境 的词向量：",model2.wv["环境"])
print("环境 词向量的形状：",model2.wv["环境"].shape)
```

```python
# 任务3：与“好吃”最接近的3词
result = model2.wv.most_similar("好吃", topn=3)
print("和【好吃】语义最接近的3个词：",result)
```

```python
# 任务4：计算与“好吃”的相似度
sim1 = model2.wv.similarity("好吃", "美味")
print("好吃 与 美味 相似度：", sim1)

sim2 = model2.wv.similarity("好吃", "蟑螂")
print("好吃 与 蟑螂 相似度：", sim2)
```

```python
# 任务5：执行向量运算
result2 = model2.wv.most_similar(
    positive=["餐厅", "聚会"],
    negative=["安静"],
    topn=1)

word, score = result2[0]
print(f"餐厅 + 聚会 - 安静 = {word}")
```
